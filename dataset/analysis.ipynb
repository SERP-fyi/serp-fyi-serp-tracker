# serp.fyi — SERP & AI Visibility Analysis
Analyze keyword ranking and AI Overview exposure data from serp.fyi campaigns.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sample.csv')
print(df.head())
```

```python
df['ai_visibility_score'].hist(bins=10)
plt.title('AI Visibility Score Distribution — serp.fyi')
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()
```

```python
print('Average AI Score:', df['ai_visibility_score'].mean())
print('AI Overview detected:', df['ai_overview_detected'].sum(), 'campaigns')
print('Perplexity mentions:', df['perplexity_mention'].sum(), 'campaigns')
```
